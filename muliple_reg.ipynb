```python
## Lets prepeare the dataframe
dataset=pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
dataset['Price']=california.target
dataset.head()
```

```python
dataset.describe()
dataset.isnull().sum()
```

```python
import seaborn as sns
sns.pairplot(dataset.corr())
```

```python
dataset.corr()
```

```python
sns.heatmap(dataset.corr(),annot=True)
```

```python
## Independent and Dependent Feature
X=dataset.iloc[:,:-1] #independent features
y=dataset.iloc[:,-1]  #Dependent feautures
from sklearn.model_selection import train_test_split
 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
## prediction for the test data 
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(rmse)
print(mae)
```

```python
## r squared and adjusted R square
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```

```python
r=1-(1-score)*(len(y_test)-1)/((len(y_test))- (X_test.shape[1]-1))
r
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel("The truth data")
plt.ylabel("the predicted data")
```
